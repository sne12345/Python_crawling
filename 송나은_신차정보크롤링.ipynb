```
import urllib.request
from bs4 import BeautifulSoup
import requests
from openpyxl import Workbook

write_wb = Workbook()  # 엑셀 파일을 열어줌
write_ws = write_wb.create_sheet()  # 시트를 만들어줌
write_ws.title = '색상'
write_ws['A1'] = '외장'
write_ws['B1'] = '내장'
write_ws['C1'] = '가격'

from bs4 import BeautifulSoup  # 파이썬에서 HTML코드를 가공해줌 # 모듈 -> 도구가방
import requests  # 서버랑 클라이언트로서 통신하게 해줌

url =  'http://www.carnoon.co.kr/newcar/estimate/10001'
i = 1
while i < 10 :
  url = url[:-1] + ("%d" %i)
  i = i + 1
  result = requests.get(url)
  result.encoding = 'utf-8'
  tem = result.text 
  soup = BeautifulSoup(tem,'html.parser')

  a = soup.find_all('div',{'class':'brand'}) # id로 찾는건 어떻게 합니까?
  if a == [] :
    pass
  else :
    a = a[1].text
    print(a[:-5])

    block1 = soup.find_all('ul',{'id':'colorExtBox'})
    if block1 == [] :
      pass
    else :
      block1 = block1[0].text
      block1 = block1.replace('\xa0','')
      block1 = block1.split('\n')
      del block1[0], block1[-1]
      for j in block1 :
        str1 = j
        print(j)
        write_ws.append([str1])
      print('\n')

  block2 = soup.find_all('ul',{'id':'colorIntBox'})
  if block2 == [] :
    pass
  else :
    block2 = block2[0].text
    block2 = block2.replace('\xa0','')
    block2 = block2.split('\n')
    del block2[0], block2[-1]
    for k in block2 :
      str2 = k
      print(k)
      write_ws.append([str2])
    print('\n')

write_wb.save('UnivBook.xlsx')
```
